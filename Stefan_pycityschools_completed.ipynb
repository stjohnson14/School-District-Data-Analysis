```python
total_schools_name = school_data_complete.groupby('School Name')
type_df = total_schools_name['type'].unique()
type_df
```

```python
total_students = school_data_complete.groupby(['School Name']).count()['student_name']
per_school_budget = school_data_complete.groupby(["School Name"]).mean()["budget"]
per_student_budget = per_school_budget / total_students
student_math_avg = school_data_complete.groupby(["School Name"]).mean()['math_score']
student_reading_avg = school_data_complete.groupby(["School Name"]).mean()['reading_score']
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
per_school_passing_math = school_passing_math.groupby(["School Name"]).count()["student_name"]
per_school_counts = school_data_complete.groupby(["School Name"]).count()["Student ID"]
per_school_passing_math = per_school_passing_math / per_school_counts * 100
school_passing_read = school_data_complete[(school_data_complete["reading_score"] >= 70)]
per_school_passing_read = school_passing_read.groupby(["School Name"]).count()["student_name"]
per_school_passing_read = per_school_passing_read / per_school_counts * 100
```

```python
school_passing_overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
per_school_passing_overall = school_passing_overall.groupby(["School Name"]).count()["student_name"]
per_school_passing_overall = per_school_passing_overall / per_school_counts * 100
per_school_passing_overall
```

```python
# Convert dictionary into a DataFrame
school_results = {
    "School Type": type_df,
    "Total Students": total_students,
    "Total School Budget": per_school_budget,
     "Budget Per Student":per_student_budget,
      "Average Math Score": student_math_avg,
     "Average Reading Score": student_reading_avg,
     "% Passing Math": per_school_passing_math,
     "% Passing Reading": per_school_passing_read,
     "% Overall Passing": per_school_passing_overall
}

# Convert dictionary into DataFrame
school_results = pd.DataFrame({"School Type": type_df,
                               "Total Students": total_students,
                               "Total School Budget": per_school_budget,
                                "Budget Per Student" : per_student_budget,
                                "Average Math Score" : student_math_avg,
                                "Average Reading Score": student_reading_avg,
                                "% Passing Math": per_school_passing_math,
                                "% Passing Reading": per_school_passing_read,
                                "% Overall Passing": per_school_passing_overall
                              })

# Reformat to dollars
school_results["Total School Budget"] = school_results["Total School Budget"].map("${:,.0f}".format)
school_results["Budget Per Student"] = school_results["Budget Per Student"].map("${:,.0f}".format)
school_results["Average Math Score"] = school_results["Average Math Score"].map("{:.0f}".format)
school_results["Average Reading Score"] = school_results["Average Reading Score"].map("{:.0f}".format)
school_results["% Passing Math"] = school_results["% Passing Math"].map("{:.2f}%".format)
school_results["% Passing Reading"] = school_results["% Passing Reading"].map("{:.2f}%".format)
school_results["% Overall Passing"] = school_results["% Overall Passing"].map("{:.2f}%".format)

school_results
```

```python
top_overall_schools = school_results.sort_values(['% Overall Passing'], ascending=False)
top_overall_schools.head()
```
